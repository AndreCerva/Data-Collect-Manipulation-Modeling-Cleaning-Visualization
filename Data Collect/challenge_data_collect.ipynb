Existen 13 funciones de agregación (metodos) incluidas en Pandas:

first(): Devuelve el primer elemento.

last(): Devuelve el último elemento.

## Manipulación de JSON Python 
JavaScript Object Notation (JSON) El formato mas utilizado para trasferir informacion (intercambio de datos a través de la web)

1) Busquemos acceder a las 'description' del json

```python
import json
otro_JSON='''
{
    "marcadores": [
      {
        "latitude": 40.416875,
        "longitude": -3.703308,
        "city": "Madrid",
        "description": "Puerta del Sol"
      },
      {
        "latitude": 40.417438,
        "longitude": -3.693363,
        "city": "Madrid",
        "description": "Paseo del Prado"
      },
      {
        "latitude": 40.407015,
        "longitude": -3.691163,
        "city": "Madrid",
        "description": "Estación de Atocha"
      }
    ]
}
'''
jsonlist=json.loads(otro_JSON)
for i in range (len(jsonlist['marcadores'])):
    print(jsonlist['marcadores'][i]['description'])
```

## Lecturas API
Haz una consulta a la siguiente API y obten todos los elementos de 'sunrise' del JSON que da como resultado

```python
import requests
API=" https://api.sunrise-sunset.org/json"
data=requests.get(API)
dataget=data.json()
print(dataget['results']['sunrise'])
df_API=pd.DataFrame(dataget)
print(df_API)
```

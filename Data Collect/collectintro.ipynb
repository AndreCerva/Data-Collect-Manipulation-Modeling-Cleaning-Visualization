## Lectura de APIs

Para poder obtener datos desde una API o servicio web podemos utilizar el modulo de Python requests. Este modulo permite realizar peticiones HTTP a una dirección URL de manera simple

```python
import requests
```

Para realizar la petición, usaremos la función get() pasando como unico argumento la url deseada. Esta función devuelve un objeto de la clase requests.models.response, devuelve un codigo que indica el estado d enuestra peticion

```python
consult=requests.get("http://api.open-notify.org/this-api-doesnt-exist")
#404 = not found 50x = server error
consult
```

En el siguiente ejemplo recibimos información en tiempo real desde la nasa consultando este mediante una API

```python
response = requests.get("http://api.open-notify.org/astros.json/")
#200 = OK
response
```

Una API por estandar devuelve los valores en formato JSON. Para obtener estos datos en formato de un diccionario de Python utilizamos el metodo Json

```python
data=response.json()
dfNASA=pd.DataFrame(data['people'])
```

Podemos obtener la información de una fila de este DataFrame mediant eel metodo loc el numero de fila y la columna

```python
dfNASA.loc[0,'name']
dfNASA.head(3)
```
